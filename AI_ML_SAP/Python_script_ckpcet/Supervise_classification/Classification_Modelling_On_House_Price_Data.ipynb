## Prepare Data for Machine Learning

```python
# Seperate the target variable and independent variable
target = df_houseprice['Sold']
```

```python
# Independent Variable
df_independent = df_houseprice.drop('Sold', axis = 1)
df_independent
```

## Seperate the Categorical Column from Independent Variable

```python
df_cat = df_independent.select_dtypes(include = "object")
df_cat.head()
```

```python
# Encode the Categorical variable
encoded = pd.get_dummies(df_cat)
encoded = encoded.reset_index()
```

```python
encoded.head()
```

## Scale the Numerical Data

```python
df_num = df_independent.select_dtypes(include = 'number')
df_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initiate the Standard Scaler
ss = StandardScaler()
```

```python
# Transform The Data 
scale = ss.fit_transform(df_num)
scale
```

```python
df_scale = pd.DataFrame(scale, columns = df_num.columns)
df_scale = df_scale.reset_index()
df_scale
```

```python
X = pd.concat([df_scale,encoded],axis = 1)
X.head()
```

```python
# Split the Dataset For Training and Testing

import statsmodels
import statsmodels.api as sm

X = sm.add_constant(X)
X.head()
```

```python
# Splitting in train and test subsets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, random_state = 10, test_size = 0.2)
```

```python
# print dimensios of train set
print('X_train', X_train.shape)
print('y_train', y_train.shape)

# print dimensions of test set
print('X_test', X_test.shape)
print('y_test', y_test.shape)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create a Decision Tree Classifier
tree_model = DecisionTreeClassifier()

# Fit the model to the Training Data
tree_model.fit(X_train,y_train)

# Predict the Training Data
predictions = tree_model.predict(X_test)
```

```python
# Check the Accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy :", accuracy)
```
