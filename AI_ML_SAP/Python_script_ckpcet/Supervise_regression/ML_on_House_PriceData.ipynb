## Prepare Data for Machine Learning

```python
# We know that the Machine Learning Alorithm are Mathemaitical Expression so we need to provide the data to the algorithm in specific format
# Encoding
```

```python
# Step 1 : Divide the detabase on their datatype
df_character = df_houseprice.select_dtypes(include = ['object'])
df_character.columns
```

```python
# Stpe 2 : Use dummy encoding to convert character columns into integer column
dummy_var = pd.get_dummies(data = df_character, drop_first = True)
dummy_var = dummy_var.reset_index()
dummy_var2 = pd.get_dummies(df_character)
```

```python
dummy_var.head()
```

```python
dummy_var2.head()
```

```python
df_number = df_houseprice.select_dtypes(include = [np.number])
df_number.head()
```

```python
from sklearn.preprocessing import StandardScaler
df_target = df_houseprice['Price']
df_number = df_number.drop('Price',axis = 1)
df_number.head()
```

```python
# Initiate the Standard Scaler
ss = StandardScaler()
```

```python
# Step 3 : Transform the data 
Scale = ss.fit_transform(df_number)
Scale
```

```python
df_scale = pd.DataFrame(Scale, columns = df_number.columns)
df_scale = df_scale.reset_index()
df_scale.head()
```

```python
# Step 4 : Make a DataFrame for independent variables
df_independent = pd.concat([df_scale, dummy_var],axis = 1)
df_independent.head()
```

```python
# Step 5 : Split the data for training  and testing
import statsmodels
import statsmodels.api as sm

df_independent = sm.add_constant(df_independent)
df_independent.head()
```

## Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_independent, df_target, random_state = 10, test_size = 0.2)
```

```python
# print dimensios of train set
print('X_train', X_train.shape)
print('y_train', y_train.shape)

# print dimensions of test set
print('X_test', X_test.shape)
print('y_test', y_test.shape)
```

```python
X_train.columns
```

### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

#Fit the model to the training data
model.fit(X_train, y_train)

# Predict on the testing data
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

# Calculate mean squared error
mse = mean_squared_error(y_test,predictions)
print("Mean Squared Error : " , mse)
```

```python
from sklearn.metrics import mean_absolute_error


# Calculate mean absolute error
mae = mean_absolute_error(y_test,predictions)
print("Mean Absolute Error : " , mae)
```

```python
from sklearn.metrics import root_mean_squared_error


# Calculate Root Mean Squared error
rmse = root_mean_squared_error(y_test,predictions)
print("Root Mean Squared Error : " , rmse)
```
